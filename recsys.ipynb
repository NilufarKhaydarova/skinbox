```python
import pandas as pd
import numpy as np
from scipy.sparse.linalg import svds
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1= pd.read_csv('Skincare Database.csv', delimiter=',')
print(df1.shape)
print(df1.info())
display(df1.head())

df_review = pd.read_csv('reviews.csv', delimiter=',',encoding= 'unicode_escape')
print(df_review.shape)
print(df_review.info())
display(df_review.head())
```

```python
df1.isnull().values.sum()
df_products = df1[['Name of the product',  'Description','Function', 'Skin Type', 'Purpose', 'Price']]
```

```python
df_products.head()
```

```python
rev = df_review.drop('Recommendation percent ',1)
rev = rev.rename(columns={'Name of the product  ':'Name of the product'})

rev.isnull().values.sum()
```

```python
df = pd.merge(rev, df_products, on='Name of the product' )
df.head()
```

```python
# Function that computes the weighted rating of each product
ave_r=df['Review grade'].mean()
min_r=df['Number of reviews '].min()
max_r= df['Number of reviews '].max()
print('Minimum and maximum number of reviews: ', min_r,',', max_r)

def weighted_rating(new_df, m=min_r, C=ave_r):
    v = new_df['Number of reviews ']
    R = new_df['Review grade']
    # Calculation based 
    return (v/(v+m) * R) + (m/(m+v) * C)

df['score'] = df.apply(weighted_rating, axis=1)
```

```python
min_common_items = 5

def predict_scores(df_sim, ser_target):
    ret = {}
    for item1 in df_sim.index:  # not yet rated by the target user
        v1 = df_sim.loc[item1]
       # if v1.isnull().sum() > 0:  # debug
           # print('v1:',v1)  # debug
        v11 = v1[ v1.notnull() ]
        t11 = ser_target[ v1.notnull() ]
        pred1 = (v11 * t11).sum() / np.abs(v11).sum()
        pred2=(v11 * t11).sum() / np.abs(v11).sum()
        #print('v11:',v11)  # debug
        #print('t11:',t11)  # debug
        print('pred1:',pred1)  # debug
        ret[item1] = pred1
    
    ser_ret = pd.Series(ret)
    
    return ser_ret.sort_values(ascending=False)
```

```python
def get_recomm_by_item_sim(df, target_dic):
    ser_target = pd.Series(target_dic)
    print(target_dic)  # debug
    print(ser_target)  # debug
    # make dataframe with columns included in target_dic
    #print(df.shape)  # debug
    df_scores = df[ ser_target.index ]
    #print(df_scores.shape)  # debug
    # drop rows included in target_dic (already rated)
    #df_scores = df_scores.drop(index=ser_target.index)
    #print(df_scores.shape)  # debug
    #display(df_scores.head())  # debug
    recomm = predict_scores(df_scores, ser_target)
    
    return recomm
```

```python
from sklearn import preprocessing
#Function
le = preprocessing.LabelEncoder()
le.fit(df['Function'])
df['Function'] = le.transform(df['Function'])
function_to_num = {k:v for v,k in enumerate(le.classes_)}

le.inverse_transform(df['Function'])
```

```python
#Skin Type
le = preprocessing.LabelEncoder()
le.fit(df['Skin Type'])
df['Skin Type'] = le.transform(df['Skin Type'])
skintype_to_num = {k:v for v,k in enumerate(le.classes_)}
le.inverse_transform(df['Skin Type'])
```

```python
#Purpose
le = preprocessing.LabelEncoder()
le.fit(df['Purpose'])
df['Purpose'] = le.transform(df['Purpose'])
purpose_to_num = {k:v for v,k in enumerate(le.classes_)}

le.inverse_transform(df['Purpose'])
```

```python
recomm = get_recomm_by_item_sim(df,
                                {'Purpose':purpose_to_num['Treat  '],  
                                 'Skin Type':skintype_to_num['Aging  '],
                                 'Function':function_to_num['Renews cells  '],
                                 'score':5})
print('Recommendation:')
print(recomm.head())
```
